```python
input_file = '/data/cooked_v3.pkl'
main_df = pd.read_pickle(input_file)
main_df = main_df
print("read")
```

```python
# TODO: move to utils
def date_part(dt):
    return str(dt).split(' ')[0]
unique_days = sorted(list(set(map(date_part , list(main_df.index.unique())))))
len(unique_days), unique_days[0:2], unique_days[-3:-1]
```

```python
main_df['CalcDateTime'] = main_df.index
main_df['Date'] = main_df['CalcDateTime'].dt.strftime("%Y-%m-%d")
pred_dates = unique_days
df_main_subset = main_df[main_df.Date.isin(pred_dates)]
```

```python
mnemonics = list(main_df['Mnemonic'].unique())
```

```python
def prepare(input_df, mnemonic):
    input_df = input_df[input_df.Mnemonic==mnemonic]
    price_df = input_df[['EndPrice']].resample('1H').mean().dropna()
    price_df['CalcDateTime'] = price_df.index
    price_df['Date'] = price_df['CalcDateTime'].dt.strftime("%Y-%m-%d")
    price_df['Date-Mnemonic'] = price_df['Date'] + '-' +  mnemonic
    price_df['Time'] = price_df['CalcDateTime'].dt.strftime("%H")
    price_df=price_df.drop(columns=['CalcDateTime'])
    mean_price_df = price_df.groupby('Date')['EndPrice'].agg([ np.mean, np.std])
    price_df=pd.merge(price_df, mean_price_df, on='Date')
    #price_df['NormEndPrice'] = (price_df['EndPrice'] -  price_df['mean']) / price_df['std']
    price_df['NormEndPrice'] = (price_df['EndPrice'] -  price_df['mean']) / price_df['mean']   
    price_df['Mnemonic'] = mnemonic
    return price_df
```

```python
a = prepare(df_main_subset, 'SIE')
a.head()
```

```python
def prepare_all(input_df, mnemonics):
    df_lst = []
    for mnemonic in mnemonics:
        df_lst.append(prepare(input_df, mnemonic))
    return pd.concat(df_lst, axis=0, sort=False)
r = prepare_all(df_main_subset, mnemonics)
r['Mnemonic'].unique()
r.shape
```

```python
def prep_for_matrix(input_df):
    def myagg(c):
        n = c[['NormEndPrice']].T
        cnames=list(c['Time'].T.values)
        n.columns = cnames
        return n
    x=input_df.groupby('Date-Mnemonic').apply(myagg)
    x.index = x.index.droplevel(1)
    return x
```

```python
M = prep_for_matrix(r)
M.fillna(0.0, inplace=True)
```

```python
from sklearn.cluster import KMeans
import numpy as np
kmeans = KMeans(n_clusters=5, random_state=466460).fit(M.values)
centers=kmeans.cluster_centers_
```

```python
fr = pd.DataFrame(centers)
fr.columns = [u'08', u'09', u'10', u'11', u'12', u'13', '14', '15', '16', '17', '18', '19', '20']
fr = fr.T
fr.columns = map(lambda n: "cluster" + str(n), fr.columns)
%matplotlib inline
mpl.rcParams['figure.figsize'] = (20, 10) # use bigger graphs
fr.plot()
```
